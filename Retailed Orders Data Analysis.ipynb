### read data from the file and handle null values

```python

import pandas as pd
df = pd.read_csv('C:\\Users\\yadav\\Documents\\SQL & Pyhton Projects\\Retail Orders\\orders.csv')
df.head(20)
```

```python
#finding distinct values available in distinct mode
df = pd.read_csv('C:\\Users\\yadav\\Documents\\SQL & Pyhton Projects\\Retail Orders\\orders.csv', na_values = ['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

### rename columns names, makes them lower case and  replace space with underscore

```python

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.head(5)
```

### derive new columns discounts, sales price and profit

```python
df['discount'] = df['list_price']*df['discount_percent']*0.01
df.head(5)
```

```python
df['sales_price'] = df['list_price'] - df['discount']
df.head(5)
```

```python
df['profit'] = df['sales_price'] - df['cost_price']
df.head(5)
```

### convert order_date data type to datetime

```python
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.dtypes
```

### drop columns cost_price, list_price, discount_percent which are of no use

```python
df.drop(columns = ['list_price','discount_percent','cost_price'], inplace = True)
```
